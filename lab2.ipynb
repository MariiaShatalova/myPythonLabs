```python
import pandas as pd

file_path = r'C:\Users\Maria\Downloads\NationalNames.csv'
df = pd.read_csv(file_path)

# Групування за іменем та статтю, підрахунок загальної кількості народжень
gender_counts = df.groupby(['Name', 'Gender'])['Count'].sum().unstack(fill_value=0)

# Відфільтрувати гендерно нейтральні імена (записи для обох статей)
gender_neutral_names = gender_counts[(gender_counts['M'] > 0) & (gender_counts['F'] > 0)]

# Знайти всі роки
all_years = df['Year'].nunique()

# Відфільтрувати імена, присутні в кожному році
names_in_all_years = gender_neutral_names[gender_neutral_names['M'] > 0].index.intersection(
    gender_neutral_names[gender_neutral_names['F'] > 0].index
)

# Підрахунок загальної кількості для гендерно нейтральних імен
popular_gender_neutral_names = df[df['Name'].isin(names_in_all_years)].groupby('Name')['Count'].sum()

print("Найпопулярніші гендерно-нейтральні імена (ті, що присутні кожного року):")
print(popular_gender_neutral_names)
```

```python
import pandas as pd

file_path = r'C:\Users\Maria\Downloads\NationalNames.csv'
df = pd.read_csv(file_path)

# Підрахунок загальної кількості народжень для кожного імені
total_counts = df.groupby('Name')['Count'].sum()

# Визначення порогу для непопулярних імен (наприклад, менше 100)
threshold = 100

# Фільтрація непопулярних імен
unpopular_names = total_counts[total_counts < threshold]

# Найпопулярніше серед непопулярних імен
most_popular_unpopular_name = unpopular_names.idxmax()
most_popular_unpopular_count = unpopular_names.max()

# Виведення результату
print(f"Найпопулярніше серед непопулярних імен: {most_popular_unpopular_name} ({most_popular_unpopular_count} народжень).")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\Maria\Downloads\NationalNames.csv'
df = pd.read_csv(file_path)

# Фільтрація даних для імен "John" та "Mary"
john_data = df[df['Name'] == 'John']
mary_data = df[df['Name'] == 'Mary']

# Групування за роком та підрахунок загальної кількості народжень для кожного імені
john_counts = john_data.groupby('Year')['Count'].sum()
mary_counts = mary_data.groupby('Year')['Count'].sum()

# Побудова графіка
plt.figure(figsize=(10, 6))

# Графік для "John"
plt.plot(john_counts.index, john_counts.values, color='blue', label='John')

# Графік для "Mary"
plt.plot(mary_counts.index, mary_counts.values, color='red', label='Mary')

# Налаштування графіка
plt.title('Розподіл імен "John" та "Mary" по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість народжень')
plt.xticks(john_counts.index[::4], rotation=45)
plt.grid()
plt.legend()

# Відображення графіка
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r'C:\Users\Maria\Downloads\NationalNames.csv'
df = pd.read_csv(file_path)

# Фільтрація даних для жіночих імен "John" та чоловічих імен "Mary"
female_john_data = df[(df['Name'] == 'John') & (df['Gender'] == 'F')]
male_mary_data = df[(df['Name'] == 'Mary') & (df['Gender'] == 'M')]

# Групування за роком та підрахунок загальної кількості народжень для кожного імені
female_john_counts = female_john_data.groupby('Year')['Count'].sum()
male_mary_counts = male_mary_data.groupby('Year')['Count'].sum()

# Побудова графіка
plt.figure(figsize=(10, 6))

# Графік для жіночого імені "John"
plt.plot(female_john_counts.index, female_john_counts.values, color='blue', label='Жіноче ім\'я "John"')

# Графік для чоловічого імені "Mary"
plt.plot(male_mary_counts.index, male_mary_counts.values, color='red', label='Чоловіче ім\'я "Mary"')

# Налаштування графіка
plt.title('Розподіл кількості імен "John" та "Mary" по роках')
plt.xlabel('Рік')
plt.ylabel('Кількість народжень')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

# Відображення графіка
plt.tight_layout()
plt.show()
```
