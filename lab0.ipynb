```python
%%HTML
<svg width="300" height="180">
  <rect width="300" height="60" style="fill:red;" />
  <rect y="60" width="300" height="60" style="fill:yellow;" />
  <rect y="120" width="300" height="60" style="fill:green;" />
  <polygon points="150,60 131,120 180,85 120,85 169,120" style="fill:black;" />
</svg>
```

```python
def is_prime(n):
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def kth_prime(k):
    if not isinstance(k, int) or k <= 0:
        return("Помилка. k повинно бути додатним цілим числом та не 1")
    
    count = 0
    num = 1
    while count < k:
        num += 1
        if is_prime(num):
            count += 1
    return num
```

```python
k = int(input("Введіть значення k: ")) 
result = kth_prime(k) 
if isinstance(result, str): 
    print(result) 
else: 
    print(f"{k}-те просте число: {result}") 
```
